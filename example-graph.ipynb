# Computation Graph Example

```python
from simplegrad.engine import Num, display_dag, relu, log, exp
```

```python
a = Num(2)
b = Num(3)
c = a - b
d = relu(c)
e = d + a
f = e * e
g = Num(5)
h = exp(g)
i = h / a
j = i + h
k = j * f
l = relu(k)
m = log(l)

a.name = 'a'
b.name = 'b'
c.name = 'c'
d.name = 'd'
e.name = 'e'
f.name = 'f'
g.name = 'g'
h.name = 'h'
i.name = 'i'
j.name = 'j'
k.name = 'k'
l.name = 'l'
m.name = 'm'
```

```python
m.backward()
display_dag(m, scale=.16, font_scale=1.8)
```
